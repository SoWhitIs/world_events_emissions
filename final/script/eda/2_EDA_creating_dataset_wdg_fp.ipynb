```python
# import libraries
import pandas as  pd
import os
```

```python
# Load dataset 
file_path ="../../resources/data/decade_co_emissions_fp.csv"
# file_dir=os.path.join("resources")
data_df = pd.read_csv(file_path)
data_df
```

#### Great Recession, 2007-2012

```python
#filter to "capture" The Great Recession years of 2007-2012
# Create a Great Recession Dataframe
recession_df=data_df.loc[(data_df['year']>= 2007) & (data_df['year'] <= 2012)]
recession_df
```

```python
recession_df.dtypes
```

```python
#subset
#write to csv file
#write final dataset to csv file for furture use, with index columns, without index
file_path ="../../resources/data/recession_fp.csv"
recession_df.to_csv(file_path, index=False)
```

#### CO2 by Type by Country

```python
#create dataframe without the following labels:
#'Africa', 'Europe', 'Asia', 'Australia', 'South America', 'North America', 'Antartica', 'Oceania', 'Caribbean', 'Central America', 'World', 'North America (excl. USA)', 'Europe (excl. EU-27)', 'Europe (excl. EU-28)', 'Asia (excl. China & India)'
# use NOT(~), (!=) operator (!=) 
# ex.df[df['column] != 'C'] , df[~df['column'].isin(['value', 'value'])]


#copy recession_df dataframe
country_df=recession_df.copy()
```

```python
remove_reg =['Africa', 'Europe', 'Asia', 'Australia', 'South America', 'North America', 'Antartica', 'Oceania', 'Caribbean', 'Central America', 'World', 'North America (excl. USA)', 'Europe (excl. EU-27)', 'Europe (excl. EU-28)', 'Asia (excl. China & India)']
country_df=country_df[~country_df['geo_reg'].isin(remove_reg)]
```
