### Alternative

```python
# alternative multiple dataframes/filter
#after creating Carribean and Central America Region
# REF: https://www.statology.org/pandas-filter-multiple-conditions/
# DataFrame.isin(values)[source]Â¶ ex.   df[df.team.isin(filter_list2)]
geoarea=['Africa', 'Europe', 'Asia', 'Australia', 'South America', 'North America', "Antartica", 'Oceania', 'Caribbean', 'Central America']
recession_df[recession_df.country.isin(geoarea)]
```

```python
#create new dataframe
region_df =recession_df[recession_df.country.isin(geoarea)]
print(region_df.shape)
```

```python
region_df.tail(10)
```

```python
#rename columns from country to continent, sources:https://datascienceparichay.com/article/pandas-rename-column-names/#:~:text=To%20rename%20columns%20of%20a%20dataframe%20you%20can,attribute%20to%20your%20new%20list%20of%20column%20names.
# ex. df = df.rename(columns={"Category":"Pet"})
region_df=region_df.rename(columns={'country':'geo_reg'})
region_df.head()
```

```python
#plot C02 by country from 2007-2012
colors=['red','green','purple','violet','peru','blue','orange','turquoise','teal']
# region_df.groupby(by="geo_reg").mean()['co2'].plot(kind="bar", color=colors)
average_region.plot(kind="bar", color=colors)
plt.ylabel("CO2(million tonnes)")
plt.xlabel("Geographical Region")
plt.title("Average of CO2 Emisions by Geographical Region, 2007-2012")
plt.xticks(rotation=45)
plt.figure(figsize=(15,5))
```

```python
#pie plot
labels = 'Africa', 'Asia', 'Australia', 'Caribbean', 'Central America', 'Europe', 'North America', 'Oceania','South America'
region_df.groupby(by="geo_reg").mean()['co2'].plot(kind="pie", colors=colors, labels=None)
plt.legend( labels, loc='best', bbox_to_anchor=(-0.1, 1.),
           fontsize=10)
plt.title("Average of CO2 Emisions by Continent, 2007-2012")
plt.ylabel("")
# explode
# region_df.groupby(by="geo_reg").mean()['co2'].plot(kind="pie", labels=labels, colors=colors)
```
