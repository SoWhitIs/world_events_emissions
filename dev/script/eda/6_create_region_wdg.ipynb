```python
# import libraries
import pandas as  pd
import matplotlib.pyplot as plt
import os

from matplotlib import style
# 
style.use('ggplot')
```

```python
# load dataset
file_path ="../../resources/data/great_recession_index.csv"
# file_dir=os.path.join("resources")
recession_df = pd.read_csv(file_path)
recession_df
```

#### The Caribbean

```python
#create a dataframe of all carribean countries:
# filter for Carribean Countries
carb=['British Virgin Islands', 'Anguilla', 'Antigua and Barbuda', 'Aruba',  'Bahamas', 'Barbados', 'Bermuda', 'British Virgin Islands', 'Cayman Islands', 'Cuba', 'Dominica', 'Dominican Republic', 'Grenada', 'Guadeloupe', 'Haiti', 'Jamaica', 'Martinique', 'Montserrat', 'Netherlands Antilles', 'Puerto Rico', 'Saint Kitts and Nevis', 'Saint Lucia', 'Saint Vincent and the Grenadines', 'Trinidad and Tobago', 'Turks and Caicos Islands', 'U.S. Virgin Islands']
caribbean_df=recession_df[recession_df.country.isin(carb)]
```

```python
#check_df
caribbean_df
```

```python
# add the sum of c02 and co2_per_capita, groupby year, create new dataframe
car_df=caribbean_df.groupby('year').sum()[['population', 'co2',
       'co2_per_capita', 'co2_per_gdp', 'co2_growth_prct', 'consumption_co2',
       'consumption_co2_per_capita', 'share_global_co2']]
car_df
```

```python
#add country/region to cen_amer_df
#pd.Dataframe[new_column] = value 
car_df['country'] = "Caribbean"
```

```python
car_df['iso_code']=" "
```

```python
car_df['record_no']=" "
```

```python
#convert index to column
car_df.reset_index(inplace=True)
```

```python
#check columns
car_df.columns
```

```python
#reorder columns
car_df=car_df[['record_no', 'iso_code', 'country', 'population','year', 'co2',
       'co2_per_capita', 'co2_per_gdp', 'co2_growth_prct', 'consumption_co2',
       'consumption_co2_per_capita', 'share_global_co2']]
```

```python
#check_dataframe
car_df
```

```python
# append end onf recession_df, with #continous index value, ignore_index=True
recession_df=recession_df.append(car_df,ignore_index=True )
```

```python
#check tail dataframe
recession_df.tail()
```

#### Central America

```python
#create a dataframe for countries in Central America
#filter for Central American Countries
ca=['Belize', 'Costa Rica', 'El Salvador', 'Guatemala', 'Honduras', 'Nicaragua', 'Panama']
central_america_df=recession_df[recession_df.country.isin(ca)]
```

```python
central_america_df.head()
```

```python
central_america_df.columns
```

```python
# add the sum of c02 and co2_per_capita,etc, per country, groupby year, create new dataframe
cen_amer_df=central_america_df.groupby(['year']).sum()[['population', 'co2',
       'co2_per_capita', 'co2_per_gdp', 'co2_growth_prct', 'consumption_co2',
       'consumption_co2_per_capita', 'share_global_co2']]
cen_amer_df
```

```python
central_america_df.groupby(['year', 'country']).sum()['co2']
```

```python
#add country/region to cen_amer_df
#pd.Dataframe[new_column] = value 
cen_amer_df['country'] = "Central America"
```

```python
cen_amer_df['iso_code']=" "
```

```python
cen_amer_df['record_no']=" "
```

```python
#convert index to column
cen_amer_df.reset_index(inplace=True)
```

```python
cen_amer_df.columns
```

```python
#reorder columns
cen_df=cen_amer_df[['record_no', 'iso_code', 'country', 'population','year', 'co2',
       'co2_per_capita', 'co2_per_gdp', 'co2_growth_prct', 'consumption_co2',
       'consumption_co2_per_capita', 'share_global_co2']]
```

```python
#check_dataframe
cen_df
```

```python
# append end of recession_df, with #continous index value, ignore_index=True
recession_df=recession_df.append(cen_df,ignore_index=True )
```

```python
recession_df.tail()
```

#### Write to new CSV file

```python
#drop record_no column from recession_df, df.drop(['B', 'C'], axis=1)
#df=df.drop(columns=['B', 'C'])
recession_df=recession_df.drop(['record_no'], axis=1)
```
