```python
# import libraries
import pandas as  pd
import matplotlib.pyplot as plt
import os
```

```python
# Load OurWorld in Data --dataset
source='https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv'
ow_data_df = pd.read_csv(source)
ow_data_df.head()
```

```python
# check for shape of dataframe, original dataset
ow_data_df.shape
```

```python
#check for data 2020-2021
ow_data_df.loc[(ow_data_df['year']== 2020 & 2021)]
```

```python
#find the earliest year
# ow_data_df.sort_values(['year'])

#find latest year
ow_data_df.sort_values(['year'], ascending= False)
```

```python
#create a copy of dataframe for wrangling
# copy original dataframe
emissions_df = ow_data_df.copy()
emissions_df.head()
```

```python
#check data typees
emissions_df.dtypes
```

```python
# Covert population column to int
# emission_df['population'] = emiss_df['population'].astype(int)
```

```python
#check columns
emissions_df.columns
```

```python
#create new dataframe with CO2 by type Emissions
co_type_df=emissions_df[['iso_code', 'country', 'year','population', 'co2','co2_per_capita','consumption_co2','consumption_co2_per_capita', 'trade_co2', 'cement_co2',
       'coal_co2', 'flaring_co2', 'gas_co2', 'oil_co2', 'other_industry_co2',
       'cement_co2_per_capita', 'coal_co2_per_capita',
       'flaring_co2_per_capita', 'gas_co2_per_capita', 'oil_co2_per_capita',
       'other_co2_per_capita']]
```

```python
# check new dataframe
co_type_df
```

```python
#filter dataframe to "capture" the years bewtween 2007-2012
co_type_df = co_type_df.loc[(co_type_df['year']>= 2007) & (co_type_df['year'] <= 2012)]
```

```python
#check dataframe
co_type_df
```

```python
# create regions for central america and the carribean, then append end to co_type_df

## filter for Carribean Countries
carb=['British Virgin Islands', 'Anguilla', 'Antigua and Barbuda', 'Aruba',  'Bahamas', 'Barbados', 'Bermuda', 'British Virgin Islands', 'Cayman Islands', 'Cuba', 'Dominica', 'Dominican Republic', 'Grenada', 'Guadeloupe', 'Haiti', 'Jamaica', 'Martinique', 'Montserrat', 'Netherlands Antilles', 'Puerto Rico', 'Saint Kitts and Nevis', 'Saint Lucia', 'Saint Vincent and the Grenadines', 'Trinidad and Tobago', 'Turks and Caicos Islands', 'U.S. Virgin Islands']
caribbean_df=co_type_df[co_type_df.country.isin(carb)]
```

```python
caribbean_df.columns
```

```python
# add the sum of c02 and co2_per_capita, groupby year, create new dataframe
car_df=caribbean_df.groupby('year').sum()[['population', 'co2', 'co2_per_capita',
       'consumption_co2', 'consumption_co2_per_capita', 'trade_co2',
       'cement_co2', 'coal_co2', 'flaring_co2', 'gas_co2', 'oil_co2',
       'other_industry_co2', 'cement_co2_per_capita', 'coal_co2_per_capita',
       'flaring_co2_per_capita', 'gas_co2_per_capita', 'oil_co2_per_capita',
       'other_co2_per_capita']]
car_df
```

```python
#add new columns
#add country/region to Carribean
#pd.Dataframe[new_column] = value 
car_df['country'] = "Caribbean"
car_df['iso_code']=" "
```

```python
#convert index to column
car_df.reset_index(inplace=True)
```

```python
#check columns
car_df.columns
```

```python
#reorder columns
car_df=car_df[['iso_code', 'country', 'year', 'population', 'co2', 
       'consumption_co2',  'trade_co2',
       'cement_co2', 'coal_co2', 'flaring_co2', 'gas_co2', 'oil_co2','other_industry_co2', 'co2_per_capita',
               'consumption_co2_per_capita',
       'other_industry_co2', 'cement_co2_per_capita', 'coal_co2_per_capita',
       'flaring_co2_per_capita', 'gas_co2_per_capita', 'oil_co2_per_capita',
       'other_co2_per_capita']]
```

```python
#check_dataframe
car_df
```

```python
#create a dataframe for countries in Central America
#filter for Central American Countries
ca=['Belize', 'Costa Rica', 'El Salvador', 'Guatemala', 'Honduras', 'Nicaragua', 'Panama']
central_america_df=co_type_df[co_type_df.country.isin(ca)]
```

```python
#check dataframe
central_america_df
```

```python
# add the sum of c02 and co2_per_capita, groupby year, create new dataframe
cen_df=central_america_df.groupby('year').sum()[['population', 'co2', 'co2_per_capita',
       'consumption_co2', 'consumption_co2_per_capita', 'trade_co2',
       'cement_co2', 'coal_co2', 'flaring_co2', 'gas_co2', 'oil_co2',
       'other_industry_co2', 'cement_co2_per_capita', 'coal_co2_per_capita',
       'flaring_co2_per_capita', 'gas_co2_per_capita', 'oil_co2_per_capita',
       'other_co2_per_capita']]
cen_df
```

```python
#add new columns
#add country/region to Central America
#pd.Dataframe[new_column] = value 
cen_df['country'] = "Central America"
cen_df['iso_code']=" "
```

```python
#convert index to column
cen_df.reset_index(inplace=True)

#check columns
cen_df.columns
```

```python
#reorder columns
cen_df=cen_df[['iso_code', 'country', 'year', 'population', 'co2', 
       'consumption_co2',  'trade_co2',
       'cement_co2', 'coal_co2', 'flaring_co2', 'gas_co2', 'oil_co2','other_industry_co2', 'co2_per_capita',
               'consumption_co2_per_capita',
       'other_industry_co2', 'cement_co2_per_capita', 'coal_co2_per_capita',
       'flaring_co2_per_capita', 'gas_co2_per_capita', 'oil_co2_per_capita',
       'other_co2_per_capita']]
cen_df.head()
```
